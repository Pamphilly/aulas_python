1- Acesso ao parque

```python
idade = int(input("Digite sua idade: "))

if idade >=12:
    tem_ingresso= input("Esta com ingresso? (sim/nao): ")
    
    if tem_ingresso== "sim":
        print("Entrada liberada")
        
    else:
        print("Precisa comprar ingresso!")

else:
    print("Idade insuficiente!")
```

2- Loguin simples

```python
nome = input("Digite seu nome: ")

if nome== "admin":
    senha_usuario = int(input("Digite sua senha:"))

    if senha_usuario == 1234:
        print("Acesso permitido")

    else:
        print("Senha incorreta")

else:
    print("Usuario invalido")
```

3- Aprovacao com frequencia

```python
nota = int(input("Digite sua nota: "))

if nota >=7:
    frequencia_aluno = int(input("Digite sua frequencia: "))
    if frequencia_aluno >=75:
        print("Aprovado")

    else:
        print("Reprovado por falta")

else:
    print("Reprovado por nota")
```

4- Numero par positivo

```python
num = int(input("Digite um numero inteiro: "))

if num >0:
    if num %2==0:
        print("Par positivo")

    else:
        print("Impar positivo")

else:
    print("Numero nao positivo")
```

6- Desconto em compra

```python
valor_compra = int(input("Digite o valor da compra: "))

if valor_compra >=100:
    tem_cupom= input("Possui cupom? (sim/nao): ")
    
    if tem_cupom == "sim":
        print("Desconto aplicado")
        
    else:
        print("Sem desconto")

else:
    print("EValor insuficiente para desconto!")
```

7- Multiplo de 2 e 3

```python
num = int(input("Digite um numero inteiro: "))

if num >0:
    if num %2==0 and num %3==0:
        print("O numero é multiplo de 2 e de 3")

    elif num %2==0:
        print("O numero é multiplo de 2")

    elif num %3==0:
        print("O numero é multiplo de 3")

else:
    print("O numero nao é positivo")
    

```

8- Cadastro com confirmacao

```python
nome = input("Digite seu nome: ")

if nome.strip ():
    idade = int(input("Qual sua idade?: "))
    
    if idade >=18:
        print("Cadastro completo")
        
    else:
        print("Cadastro incompleto: menor de idade")

elif not nome.strip():
    print("Nome obrigatório")
```

9- Triagem hospitalar simples

```python
idade = int(input("Digite sua idade: "))

if idade >=60:
    dor_forte = input("Está com dor forte? (sim/nao): ")
    
    if dor_forte == "sim":
        print("Atendimento prioritário imediato!")
        
    else:
        print("Aguardar triagem")

else:
    print("Atendimento comum")
```

10- Sistema de acesso por niveis

```python
usuario = input("Digite o usuario: ")

if usuario == "professor":
    senha_usuario = input("Digite sua senha: ")

    if senha_usuario == "python123":
        nivel_de_acesso = int(input("Digite o nivel de acesso: "))

    if nivel_de_acesso == 1:
        print("Acesso básico")
    
    elif senha_usuario != "python123":
        print("Senha inválida")

    else:
        print("Acesso avançado")

elif usuario != "professor":
    print("Usuario nao autorizado")
```
