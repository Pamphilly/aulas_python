Crie um programa que:
1. Peça a idade do usuário.
2. Mostre na tela:
▶ Criança (se for menor de 13 anos)
▶ Adolescente (de 13 a 17 anos)
▶ Adulto (18 anos ou mais)

```python
idade= int(input("Qual sua idade? "))

if idade<=13:
    print("Voce é Crianca")

elif idade<=17:
    print("Voce é Adolescente")

else:
    print("Voce é Adulto")
```
