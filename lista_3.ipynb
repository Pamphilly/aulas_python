10) Peça um número e informe se ele é:
▶ múltiplo de 2
▶ múltiplo de 3
▶ múltiplo de 5
▶ nenhum dos anteriores

```python
number=int(input("Digite um numero: "))

if number %2==0 and number %3==0:
    print("O numero é multiplo de 2 e multiplo de 3")

elif number %2==0:
    print("O numero é multiplo de 2")

elif number %3==0:
    print("O numero é multiplo de 3")

elif number %5==0:
    print("O numero é multiplo de 5")

else:
    print("Nenhum dos anteriores")
```
